```python
import os
import pandas as pd
import torch.nn as nn
```

```python
os.listdir("../Data_Sets/COMPAS")
```

```python
data_name = "compas"
# data_name = "German-Credit"

n_starting_instances = 1200
compas_dict = {
        "data_path": '../Data_Sets/COMPAS/',
        "filename_train": 'compas-train.csv',
        "filename_test": 'compas-test.csv',
        "label": "risk",
        "task": "classification",
        "lr": 1e-3,
        "d": 6,
        "H1": 10,
        "H2": 10,
        "activFun": nn.Softplus(),
        "n_starting_instances": n_starting_instances
    }

german_dict = {
        "data_path": '../Data_Sets/German_Credit_Data/',
        "filename_train": 'german-train.csv',
        "filename_test": 'german-test.csv',
        "label": "credit-risk",
        "task": "classification",
        "lr": 1e-3,
        "d": 6,
        "H1": 10,
        "H2": 10,
        "activFun": nn.Softplus(),
        "n_starting_instances": n_starting_instances
    
    }

adult_dict = {
        "data_path": "../Data_Sets/Adult/",
        "filename_train": 'adult-train.csv',
        "filename_test": 'adult-test.csv',
        "label": 'income',
        "task": "classification",
        "lr": 1e-3,
        "d": 6,
        "H1": 25,
        "H2": 25,
        "activFun": nn.Softplus(),
        "n_starting_instances": n_starting_instances
    }


data_meta_dictionaries = {
        "compas": compas_dict, 
        "adult": adult_dict, 
        "german":german_dict
    }
```

```python
train_df = pd.read_csv(data_meta_dictionaries[data_name]["data_path"]+data_meta_dictionaries[data_name]["filename_train"] )  

# test_df = pd.read_csv("../Data_Sets/COMPAS/compas-train.csv")
```

```python
# del train_df["Unnamed: 0"]
```

```python
# train_df.to_csv("../Data_Sets/COMPAS/compas-train.csv", index = False)
```

```python
train_df_0 = train_df[train_df[data_meta_dictionaries[data_name]["label"]] == 0]

train_df_1 = train_df[train_df[data_meta_dictionaries[data_name]["label"]] == 1]
```

```python
train_df = pd.concat([train_df_0, train_df_0, train_df_1])

```

```python
old_path = data_meta_dictionaries[data_name]["data_path"]+data_meta_dictionaries[data_name]["filename_train"]
new_train_path =  old_path.split(".csv")[0]+"-upsampled"+".csv"
train_df.to_csv(new_train_path, index = False)

```

```python
test_df = pd.read_csv(data_meta_dictionaries[data_name]["data_path"]+data_meta_dictionaries[data_name]["filename_test"] )  

test_df_0 = test_df[test_df[data_meta_dictionaries[data_name]["label"]] == 0]

test_df_1 = test_df[test_df[data_meta_dictionaries[data_name]["label"]] == 1]

test_df = pd.concat([test_df_0, test_df_0, test_df_1])

old_path = data_meta_dictionaries[data_name]["data_path"]+data_meta_dictionaries[data_name]["filename_test"]
new_test_path =  old_path.split(".csv")[0]+"-upsampled"+".csv"
test_df.to_csv(new_test_path, index = False)
```
